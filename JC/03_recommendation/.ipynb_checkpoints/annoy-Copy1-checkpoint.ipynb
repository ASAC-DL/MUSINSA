```python
import os
import glob
import tensorflow as tf
import tensorflow_hub as hub
import annoy
```

```python
# AnnoyIndex 설정
dimension = 128  # feature 벡터의 크기
n_trees = 10  # AnnoyIndex를 위한 트리의 개수
index = annoy.AnnoyIndex(dimension, 'euclidean')

# TensorFlow 모듈 로드
module_url = 'https://tfhub.dev/google/imagenet/mobilenet_v2_100_128/feature_vector/5'
module = hub.load(module_url)

```

```python
path = r"C:\Users\NT550009\Desktop\detail_image_aggregated"
os.chdir(path)
files = os.listdir(path)
files.reverse()  # 파일 리스트를 역순으로 정렬
img_paths = []  # 이미지 파일 경로를 저장할 리스트

for i, file in enumerate(files):
    image_path = os.path.join(path, file)  # 파일 속 이미지 파일 경로
    img_paths.append(image_path)
```

```python
# 이미지 feature 추출 및 AnnoyIndex에 추가
for i, file in enumerate(img_paths):
    # 이미지 로드
    image = tf.keras.preprocessing.image.load_img(file, target_size=(128, 128))
    image = tf.keras.preprocessing.image.img_to_array(image)
    image = image / 255.0  # 이미지를 0-1 범위로 정규화

    # 이미지 feature 추출
    feature_vector = module([image])[0]

    # AnnoyIndex에 feature 벡터 추가
    index.add_item(i, feature_vector)
```

```python
# 유사도 계산에는 MobileNet V2 모델이 사용
# AnnoyIndex는 유클리디안 거리를 기반

# AnnoyIndex 빌드
index.build(n_trees)

# 특정 query 이미지의 feature 추출 및 유사한 이웃 검색
query_image_path = "C:\Users\NT550009\Desktop\recommendation\cropped_image_2.jpg"  # 쿼리 이미지 경로 설정
query_image = tf.keras.preprocessing.image.load_img(query_image_path, target_size=(128, 128))
query_image = tf.keras.preprocessing.image.img_to_array(query_image)
query_image = query_image / 255.0  # 이미지를 0-1 범위로 정규화

query_feature_vector = module([query_image])[0]

n_neighbors = 5  # 반환할 유사한 이웃의 개수
neighbor_indices = index.get_nns_by_vector(query_feature_vector, n_neighbors)

# 결과 출력
print("Query 이미지와 유사한 이웃:")
for neighbor_index in neighbor_indices:
    neighbor_image_path = jpg_files[neighbor_index]
    print(neighbor_image_path)
```
